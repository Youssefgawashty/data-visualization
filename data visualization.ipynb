```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data.csv')

df.head()
```

```python
plt.plot([1,3,4,2,1])
plt.show()
```

```python
plt.scatter(x=df['math score'],y=df['writing score'],color= 'red')
plt.xlabel('score in math')
plt.ylabel('score in writing')
plt.title('scatter plots between two scores')
plt.show()
```

```python
plt.hist(df['reading score'])
```

```python
plt.hist(df['reading score'],bins =5)
plt.show()
```
